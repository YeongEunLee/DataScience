```python
import pandas as pd
df_raw = pd.read_excel('건국대학교 입학전형 합격자 종단연구(11차년도) 연구진전달.xlsx', sheet_name = 0)
df_final = pd.read_excel('가번호_category_계열_전형.xlsx')

#df_final은 보내준 엑셀파일(클러스터 적혀있는거)
df_merge = pd.merge(df_raw, df_final, left_on='가번호', right_on='가번호', how='inner')
```

```python
df_merge
```

```python
#문과인지 이과인지
과_lst = []

for i in range(len(df_merge['계열'])):
    if '자연1' in df_merge['계열'][i]:
        과_lst.append('이과')
    elif '자연2' in df_merge['계열'][i]:
        과_lst.append('이과')
    elif '인문1' in df_merge['계열'][i]:
        과_lst.append('문과') 
    elif '인문2' in df_merge['계열'][i]:
        과_lst.append('문과')
    elif '예체능' in df_merge['계열'][i]:
        과_lst.append('예체능')
print(len(과_lst))
```

```python
df_merge['과'] = 과_lst
```

```python
#전형별로 보기 위해 나눔
df_교과 = df_merge[df_merge['전형']=='KU교과우수자']
```
