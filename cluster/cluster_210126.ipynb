```python
import pandas as pd
df_raw = pd.read_excel('cluster_남여_고교유형_지역.xlsx', sheet_name = 1)
df_고교지역 = pd.read_excel('11data_고교지역.xlsx', sheet_name = 0)

df_merge = pd.merge(df_raw, df_고교지역, left_on='가번호', right_on='가번호', how='inner')
```

```python
df_merge
```

```python
df_지역 = df_merge.groupby('고교지역').size()
df_지역
```

```python
df_서울 = df_merge['고교지역'] == '서울'
df_경기 = df_merge['고교지역'] == '경기'

df_서울제외 = df_merge[~df_서울]
df_지방 = df_서울제외[~df_경기]
df_지방
```

```python
df_수도권 = df_merge[df_서울 | df_경기] 
df_수도권
```

```python
df_지방_group = df_지방.groupby('category').size()
df_지방_group
```

```python
df_수도권_group = df_수도권.groupby('category').size()
df_수도권_group
```

```python
df_여 = df_merge[df_merge['성별'] == '여']
df_남 = df_merge[df_merge['성별'] == '남']
```

```python
df_여_group = df_여.groupby('category').size()
df_여_group
```

```python
df_남_group = df_남.groupby('category').size()
df_남_group
```

```python
df_일반고 = df_merge['고교유형'] == '일반계고'
df_경기 = df_merge['고교지역'] == '경기'

df_일반계고 = df_merge[df_일반고]
```

```python
df_일반계고
```

```python
df_일반고외 = df_merge[~df_일반고]
df_일반고외
```

```python
df_일반계고_group = df_일반계고.groupby('category').size()
df_일반계고_group
```
