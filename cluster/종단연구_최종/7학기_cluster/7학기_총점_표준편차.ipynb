```python
import pandas as pd
df_raw = pd.read_excel('11data.xlsx', sheet_name = 0)
df_final = pd.read_excel('7개학기_가번호,계열, 유형,전형.xlsx')

#df_final은 보내준 엑셀파일(클러스터 적혀있는거)
df_merge = pd.merge(df_raw, df_final, left_on='가번호', right_on='가번호', how='inner')
```

```python
df_merge
```

```python
#문과인지 이과인지
과_lst = []

for i in range(len(df_merge['계열'])):
    if '자연1' in df_merge['계열'][i]:
        과_lst.append('이과')
    elif '자연2' in df_merge['계열'][i]:
        과_lst.append('이과')
    elif '인문1' in df_merge['계열'][i]:
        과_lst.append('문과') 
    elif '인문2' in df_merge['계열'][i]:
        과_lst.append('문과')
    elif '예체능' in df_merge['계열'][i]:
        과_lst.append('예체능')
print(len(과_lst))
```

```python
df_merge['category'] = df_merge['category'].replace([1, 2, 3, 4, 5], ['유형1', '유형2', '유형3', '유형4', '유형5'])
df_merge.rename(columns = {"category":"유형"}, inplace=True)
df_merge
```

```python
df_merge['과'] = 과_lst
```

```python
#전형별로 보기 위해 나눔
df_교과 = df_merge[df_merge['전형']=='KU교과우수자']
```

```python
df_자기추천 = df_merge[df_merge['전형']=='KU자기추천']
df_논술우수자 = df_merge[df_merge['전형']=='KU논술우수자']
df_일반생전형 = df_merge[df_merge['전형']=='수능']
df_기타 = df_merge[df_merge['전형']=='기타']
df_학교추천 = df_merge[df_merge['전형']=='KU학교추천']
```

```python
#자기추천 총점
result = df_자기추천[['총점','과', '유형']].groupby(['유형', '과']).agg(['mean'])
result
```
