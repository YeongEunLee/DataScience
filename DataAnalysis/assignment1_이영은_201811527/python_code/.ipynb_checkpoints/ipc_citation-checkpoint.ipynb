```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import pymysql
from collections import Counter
```

```python
conn = pymysql.connect(host='chamaeleon.konkuk.ac.kr',
                       port = 3306,
                       user = "class_da",
                       password = "dataanalytics",
                       db = "tp_class_da")
cursor = conn.cursor()
cursor.execute("SELECT pa.특허등록번호, ci.인용특허등록연도 FROM tp_class_da.us_patent AS pa INNER JOIN tp_class_da.citation AS ci ON pa.특허등록번호 = ci.특허등록번호 WHERE ci.인용특허등록연도 > 1900 AND pa.특허의후방인용수 > 10 GROUP BY pa.특허등록번호")
db = cursor.fetchall()
```

```python
pt_number = []
for pt_num in db:
    pt_number.append(pt_num)
```

```python
cursor1 = conn.cursor()
cursor1.execute("SELECT ci.특허등록번호 as pt_num, ci.인용특허등록연도 as ci_year FROM tp_class_da.citation AS ci WHERE ci.인용특허등록연도 > 1900")
db_ci = cursor1.fetchall()
```

```python
ci_year_list = []
for pt_num, ci_year in db_ci:
    ci_year_list.append((pt_num, ci_year))

ci_year_dict = {}
for pt_num, ci_year in ci_year_list:
    if ci_year_dict.get(pt_num):
        ci_year_dict[pt_num].append(ci_year)
    else:
        ci_year_dict[pt_num] = [ci_year]
```
