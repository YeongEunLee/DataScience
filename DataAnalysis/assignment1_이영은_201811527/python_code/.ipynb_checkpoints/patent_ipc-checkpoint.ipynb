```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import pymysql
from collections import Counter
```

```python
conn = pymysql.connect(host='chamaeleon.konkuk.ac.kr',
                       port = 3306,
                       user = "class_da",
                       password = "dataanalytics",
                       db = "tp_class_da")
cursor = conn.cursor()
cursor.execute("SELECT ip.특허등록번호 as pt_num, ip.보유IPC전체코드 as ipc_code FROM ipc ip")
db = cursor.fetchall()

cursor_ci = conn.cursor()
cursor_ci.execute("SELECT ci.특허등록번호 as pt_num, ci.인용특허등록연도 as ci_year FROM citation ci")
db_ci = cursor_ci.fetchall()
```

```python
ipc_list = []

for pt_num, ipc_code in db:
    ipc_list.append((pt_num, ipc_code[0]))

ipc_dict = {}

for num, ipc_code in ipc_list:
    if ipc_dict.get(num):
        ipc_dict[num].append(ipc_code)
    else :
        ipc_dict[num]=[ipc_code]
```

```python
ipc_dict
```

```python
pt_num_list = list(ipc_dict.keys())

df = DataFrame(columns = ["A", "B", "C", "D", "E", "F", "G", "H"], index = pt_num_list)

for i in pt_num_list:
    for j in ["A", "B", "C", "D", "E", "F", "G", "H"]:
        df.at[i, j] = Counter(ipc_dict[i])[j]
```

```python
ci_year_list = []

for pt_num, ci_year in db_ci:
    ci_year_list.append((pt_num, ci_year))
    pt_num_list.append(pt_num)
    

ci_dict = {}
for pt_num, ci_year in ci_year_list:
    if ci_dict.get(pt_num):
        ci_dict[pt_num].append(ci_year)
    else :
        ci_dict[pt_num]=[ci_year]
```

```python
filter_ci_dict = {}
for i in ci_dict.keys():
    ci = ci_dict[i]
    filter_list = []
    for j in ci:
        if j < 1900:
            filter_list.append(1800)
        elif j < 1910:
            filter_list.append(1900)
        elif j < 1920:
            filter_list.append(1910)
        elif j < 1930:
            filter_list.append(1920)
        elif j < 1940:
            filter_list.append(1930)
        elif j < 1950:
            filter_list.append(1940)
        elif j < 1960:
            filter_list.append(1950)
        elif j < 1970:
            filter_list.append(1960)
        elif j < 1980:
            filter_list.append(1970)
        elif j < 1990:
            filter_list.append(1980)
        elif j < 2000:
            filter_list.append(1990)
        elif j < 2010:
            filter_list.append(2000)
        else:
            filter_list.append(2010)
    filter_ci_dict[i] = filter_list
    
pt_num_ci_list = list(set(pt_num_list))
```
